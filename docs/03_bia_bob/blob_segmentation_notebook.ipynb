```python
import czifile
```

```python
from pathlib import Path
```

```python
import stackview
```

```python
import numpy as np
```

```python
from skimage.segmentation import watershed
```

```python
from skimage.filters import gaussian
```

Load the image data/blobs.tif

```python
image = czifile.imread(Path('data/blobs.tif'))stackview.insight(image)
```

Segment bright objects

```python
markers = np.zeros_like(image)markers[image < 30] = 1markers[image > 150] = 2elevation_map = gaussian(image, sigma=5)segmentation = watershed(elevation_map, markers)stackview.insight(segmentation)
```

Show the resulting label image

```python
stackview.insight(segmentation)
```
